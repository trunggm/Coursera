# define the text_counter function

```python
def text_counter(reviewString):
    reviewString = reviewString.lower()
    wordInString = reviewString.split();
    wordCount = {}
    for word in wordInString:
        word = word.replace(',','')
        word = word.replace('.', '')
        if word in selected_words:
            wordCount[word]=wordCount.get(word, 0)+1
    for word in selected_words:
        if not word in wordCount:
            wordCount[word] = 0
    return wordCount  
```

# count amount each of seleted_words in each of practiceProduct review

```python
practiceProduct['word_count'] = practiceProduct['review'].apply(lambda x: text_counter(x))   
```

# count total amount each of selected_words in data set and print max and min word

```python
import operator
totalSelectedWords = {}
for word in selected_words:
    totalSelectedWords[word] = sum([x[word] for x in practiceProduct['word_count']])

print "max word ", max(totalSelectedWords.iteritems(), key=operator.itemgetter(1))[0]
print "min word ", min(totalSelectedWords.iteritems(), key=operator.itemgetter(1))[0]
```
